```python
train_data = pd.read_csv('train.csv')
train_data.drop(['indicative_price','opc_scheme','original_reg_date','lifespan','listing_id'],axis=1,inplace=True)
train_data['fuel_type'].fillna('Non', inplace=True)
# train_data['fuel_type'] = train_data['fuel_type'].dropna()
x = train_data['curb_weight']
y = train_data['power']
labels = train_data['fuel_type']
print(labels.unique())
for i in labels.unique():
    print(i,labels[labels==i].count())
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)


unique_labels = label_encoder.classes_
cmap = plt.get_cmap('viridis', len(unique_labels))
for i, label in enumerate(unique_labels):
    if(label!='Non'):
        plt.scatter(x[encoded_labels == i], y[encoded_labels == i], 
                color=cmap(i), label=label, s=5)
plt.legend(title='Fuel Type')
plt.colorbar(label='Label (from third column)')
plt.xlabel('power')
plt.ylabel('engine_cap')


plt.show()
```
